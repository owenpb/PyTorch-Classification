# Recognizing digits from MNIST dataset: building NN in PyTorch

Here we will build a neural network to recognize digits from $0$ to $9$ from the <b>MNIST dataset</b>, which consists of $28 \times 28$ greyscale images. First we will construct an inital model with PyTorch, using two hidden layers followed by a $10$-unit output layer with softmax. Then we will perform hyperparameter tuning, and decide on good values for the learning rate $\alpha$, number of hidden units, weight decay (L2 penalty), and dropout $p$. 

We will see how implementing these variance reduction techniques improves test-set performance compared to an initial baseline model.

```python
import torch
import torchvision
import torch.nn as nn
from torch.utils.data import DataLoader, Subset, random_split
from torchvision import transforms

import matplotlib.pyplot as plt
import numpy as np
```

## Load dataset and prepare data loaders:

```python
training_set, validation_set = random_split(mnist_train_dataset, [55000, 5000], generator=torch.Generator().manual_seed(1))
```

## Initial model: building a neural network in PyTorch

First let's train an initial neural network where we set the number of hidden units to 32 and 16, and use learning_rate = 0.001.
Later we will tune these values and explore using dropout and weight decay.

```python
image_size = mnist_train_dataset[0][0].shape     # (1, 28, 28)
input_size = image_size[0] * image_size[1] * image_size[2]   # number of pixels

# 2 Hidden layers: first has 32 units, second has 16 units
# Output layer should have 10 units (labels are integers 0-9)
# Using CrossEntropyLoss for loss function which includes softmax activation for output layer

initial_model = nn.Sequential(

    nn.Flatten(),
    nn.Linear(input_size, 32),
    nn.ReLU(),
    nn.Linear(32, 16),
    nn.ReLU(),
    nn.Linear(16, 10)

)
```

```python
loss_fn = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(initial_model.parameters(), lr=0.001)
num_epochs = 20
torch.manual_seed(1)
```

```python
def train(model, num_epochs, train_dl, valid_dl):

    loss_hist_train = [0] * num_epochs
    accuracy_hist_train = [0] * num_epochs
    loss_hist_valid = [0] * num_epochs
    accuracy_hist_valid = [0] * num_epochs
    
    for epoch in range(num_epochs):

        for x_batch, y_batch in train_dl:

            pred = model(x_batch)
            loss = loss_fn(pred, y_batch)
            loss.backward()
            optimizer.step()
            optimizer.zero_grad()

            loss_hist_train[epoch] += loss.item()*y_batch.size(0)
            is_correct = (torch.argmax(pred, dim=1) == y_batch).float()

            accuracy_hist_train[epoch] += is_correct.sum()

        loss_hist_train[epoch] /= len(train_dl.dataset)
        accuracy_hist_train[epoch] /= len(train_dl.dataset)
        
        
        model.eval()
        with torch.no_grad():
            for x_batch, y_batch in valid_dl:

                pred = model(x_batch)
                loss = loss_fn(pred, y_batch)
                loss_hist_valid[epoch] += loss.item()*y_batch.size(0)
                is_correct = (torch.argmax(pred, dim=1) == y_batch).float()
                accuracy_hist_valid[epoch] += is_correct.sum()
                
        loss_hist_valid[epoch] /= len(valid_dl.dataset)
        accuracy_hist_valid[epoch] /= len(valid_dl.dataset)
                
        print(f'Epoch {epoch + 1}    Train Accuracy: {accuracy_hist_train[epoch]:.4f}    Validation Accuracy: {accuracy_hist_valid[epoch]:.4f}     Train Loss: {loss_hist_train[epoch]:.4f}    Validation Loss: {loss_hist_valid[epoch]:.4f} ')
        
    return loss_hist_train, loss_hist_valid, accuracy_hist_train, accuracy_hist_valid
```

```python
initial_hist = train(initial_model, num_epochs, train_dl, valid_dl)
```

## Evaluate initial model on test set

```python
pred = initial_model(mnist_test_dataset.data / 255.)
is_correct = (torch.argmax(pred, dim=1) == mnist_test_dataset.targets).float()
initial_test_accuracy = is_correct.mean()
print(f'Test accuracy: {initial_test_accuracy:.4f}')
```

Our initial baseline model has a classification accuracy of $96.57\%$ on the test dataset. However, we can also experiment with adjusting the learning rate $\alpha$, and the number of units in the hidden layers of the network. We can also try variance reduction techniques (i.e. adding dropout and using regularization) to reduce the gap between the training loss and validation loss for our model. 

These are all hyperparameters one can tune to improve the performance of our model, as we will explore below.

## Hyperparameter tuning

First let's perform a grid search over learning rate $\alpha$ and the number of units in our two hidden layers.

```python
lr_values = [0.0001, 0.001, 0.01]
units = [(512, 256), (256, 128), (128, 64), (64, 32), (32, 16)]

results = np.empty(shape=(3,5), dtype=object)
i = 0

for lr in lr_values:
    
    j = 0
    
    for unit in units:
        
        print('\n\n\n')
        print(f'lr = {lr}, hidden units = {unit}:  \n')
        
        hd1 = unit[0]
        hd2 = unit[1]
        
        model = nn.Sequential(
                    nn.Flatten(),
                    nn.Linear(input_size, hd1),
                    nn.ReLU(),
                    nn.Linear(hd1, hd2),
                    nn.ReLU(),
                    nn.Linear(hd2, 10))
        
        loss_fn = nn.CrossEntropyLoss()
        optimizer = torch.optim.Adam(model.parameters(), lr=lr)
        num_epochs = 25
        
        params = (lr, unit)
        hist = train(model, num_epochs, train_dl, valid_dl)
        
        results[i, j] = [params, hist]
        
        j += 1
        
    i += 1
```

Let's plot both the training loss and validation loss (versus epoch) for the different values of learning rate $\alpha$ and hidden units we have chosen:  

```python
fig, axs = plt.subplots(5, 3, figsize=(12, 15))


for i in range(5):
    for j in range(3):
        
        axs[i, j].plot(np.arange(num_epochs) + 1, results[j, i][1][0], '-o', label='Train loss')
        axs[i, j].plot(np.arange(num_epochs) + 1, results[j, i][1][1], '--v', label='Valid loss')
        axs[i, j].set_title(r'$\alpha = $ ' + str(results[j, i][0][0]) + ',  Hidden dims: ' + str(results[j, i][0][1]))
        axs[i, j].set_ylim(bottom=0)
        axs[i, j].legend(frameon=True, edgecolor='k')
        axs[i, j].grid(alpha=0.1)

plt.tight_layout()
plt.show()
```

For intermediate $\alpha=0.001$ we can attempt to reduce the difference between the training loss and validation loss through variance reduction techniques, such as regularization and dropout. Let's try fixing $\alpha=0.001$, and using hidden dimensions of $128$ and $64$.

We will perform another grid search, this time for different values of weight_decay (L2 penalty) and dropout probability $p$, where we now have two dropout layers in our network.

```python
lr = 0.001
hd1 = 128
hd2 = 64

weight_decay_vals = [0, 0.0001, 0.001, 0.01]
dropout_vals = [0, 0.1, 0.2, 0.5]

wd_results = np.empty(shape=(4,4), dtype=object)

w = 0

for weight_decay in weight_decay_vals:

    d = 0

    for dropout in dropout_vals:

        print('\n\n\n')
        print(f'learning rate = {lr}, weight decay = {weight_decay}, dropout = {dropout}:  \n')

        model = nn.Sequential(
                    nn.Flatten(),
                    nn.Linear(input_size, hd1),
                    nn.ReLU(),
                    nn.Dropout(dropout),
                    nn.Linear(hd1, hd2),
                    nn.ReLU(),
                    nn.Dropout(dropout),
                    nn.Linear(hd2, 10))

        loss_fn = nn.CrossEntropyLoss()
        optimizer = torch.optim.Adam(model.parameters(), lr=lr, weight_decay=weight_decay)
        num_epochs = 20

        params = (weight_decay, dropout)
        hist = train(model, num_epochs, train_dl, valid_dl)

        wd_results[w, d] = [params, hist]

        d += 1

    w += 1
```

```python
fig, axs = plt.subplots(4, 4, figsize=(12, 12))


for i in range(4):
    for j in range(4):
        
        axs[i, j].plot(np.arange(num_epochs) + 1, wd_results[i, j][1][0], '-o', label='Train loss')
        axs[i, j].plot(np.arange(num_epochs) + 1, wd_results[i, j][1][1], '--v', label='Valid loss')
        axs[i, j].set_title('Weight decay: ' + str(wd_results[i,j][0][0]) + ',  p = ' + str(wd_results[i, j][0][1]))
        axs[i, j].legend(frameon=True, edgecolor='k')
        axs[i, j].set_ylim([0, 0.7])
        axs[i, j].set_xlim(left=0)
        axs[i, j].grid(alpha=0.1)

plt.tight_layout()
plt.show()
```

With $\alpha = 0.001$, using no weight decay appears to overfit the training set. It is therefore best to set weight decay to a non-zero value; choosing $0.001$ appears to be a reasonable choice.

Instead of plotting the loss vs. epoch, let's display the accuracy of our network in classifying MNIST images. Again we can plot both training set accuracy and validation set accuracy vs. epoch.

```python
fig, axs = plt.subplots(4, 4, figsize=(11, 11))


for i in range(4):
    for j in range(4):
        
        axs[i, j].plot(np.arange(num_epochs) + 1, wd_results[i, j][1][2], '-o', color='g', label='Train accuracy')
        axs[i, j].plot(np.arange(num_epochs) + 1, wd_results[i, j][1][3], '--v', color='m', label='Valid accuracy')
        axs[i, j].set_title('Weight decay: ' + str(wd_results[i,j][0][0]) + ',  p = ' + str(wd_results[i, j][0][1]), fontsize=9.5)
        axs[i, j].legend(frameon=True, edgecolor='k', loc='lower right')
        axs[i, j].set_ylim([0.8, 1])
        axs[i, j].set_xticks([0, 5, 10, 15, 20])
        axs[i, j].set_xlim(left=0)
        axs[i, j].grid(alpha=0.1)

fig.subplots_adjust(wspace=1.5)
fig.subplots_adjust(hspace=20)

plt.tight_layout()
plt.show()
```

## Selecting a final model

For our final model, let's choose a weight decay of $0.001$ and a dropout probability of $p=0.1$, which gave a high validation accuracy (relative to other choices of these parameters) of $0.9744$, i.e. $97.44\%$ of MNIST images in our validation set were correctly classified. With $\alpha=0.001$, setting the weight decay to at least $0.001$ appears to mitigate overfitting to the training set, in this case.

```python
lr = 0.001
hd1 = 128
hd2 = 64
dropout = 0.1
weight_decay = 0.001

final_model = nn.Sequential(
            nn.Flatten(),
            nn.Linear(input_size, hd1),
            nn.ReLU(),
            nn.Dropout(dropout),
            nn.Linear(hd1, hd2),
            nn.ReLU(),
            nn.Dropout(dropout),
            nn.Linear(hd2, 10))

loss_fn = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(final_model.parameters(), lr=lr, weight_decay=weight_decay)
num_epochs = 20

final_hist = train(final_model, num_epochs, train_dl, valid_dl)
```

## Evaluate final model on test set

```python
pred = final_model(mnist_test_dataset.data / 255.)
is_correct = (torch.argmax(pred, dim=1) == mnist_test_dataset.targets).float()
final_test_accuracy = is_correct.mean()
print(f'Test accuracy: {final_test_accuracy:.4f}')
```

We can see that tuning our hyperparameters has improved accuracy on our final test set from $96.57\%$ (initial base model) to $97.86\%$ (final model after hyperparameter tuning)!
